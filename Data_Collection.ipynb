```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
stocks = [
    "AAPL","MSFT","NVDA","GOOGL","AMZN",
    "JPM","PNC","GS","MS","BLK",
    "UNH","JNJ","PFE","MRK",
    "XOM","CVX","CAT","KO","PG","BA"
]

bonds = ["UTEN","AGG","HYG","TIP","EMB"]

tickers = stocks + bonds

data = yf.download(
    tickers=tickers,
    start="2021-04-01",
    end="2026-04-02",
    auto_adjust=True
)

close_prices = data["Close"]
close_prices = close_prices[tickers]

returns = close_prices.pct_change().dropna()

gspc = yf.download(
    "^GSPC",
    start="2021-04-01",
    end="2026-04-02",
    auto_adjust=True
)["Close"]

gspc_returns = gspc.pct_change().dropna().squeeze()

returns_aligned, gspc_returns_aligned = returns.align(
    gspc_returns,
    join="inner",
    axis=0
)

gspc_returns_aligned = gspc_returns_aligned.squeeze()

market_var = gspc_returns_aligned.var()

betas = returns_aligned.apply(
    lambda asset_returns: asset_returns.cov(gspc_returns_aligned) / market_var
)

mu_daily = returns.mean()
mu_annual = returns.mean() * 252

print(mu_annual)
print(betas)
```

```python
summary = pd.DataFrame({
    "Expected Return": mu_annual,
    "Beta": betas
})
summary.to_excel("returns_and_betas.xlsx")
from google.colab import files
files.download("returns_and_betas.xlsx")
```
